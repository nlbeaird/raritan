```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as sio
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
coast = sio.loadmat("../coastlines/MAB_COAST_JOINED.mat",squeeze_me=True, struct_as_record=False);

coastlines_lon = coast['codat'][:,0]

coastlines_lat = coast['codat'][:,1]
```

```python
data_crs = ccrs.PlateCarree()
projection = ccrs.Mercator()

fig, ax = plt.subplots(
    figsize=(6.5, 6.5),
    subplot_kw=dict(projection=projection)
)

ax.set_extent([-74.464228, -74.0, 40.409759, 40.6])

plt.plot(coastlines_lon, coastlines_lat, transform=data_crs)

ax.gridlines()
```
